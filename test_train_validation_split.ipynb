```python
url="https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
```

```python
names=['sepel-length', 'sepel-width', 'petal-length', 'patel-width', 'Class']
```

```python
dataset = pd.read_csv(url, names=names)
```

```python
dataset.head()
```

```python
X=dataset.iloc[:,:4]
```

```python
X.head()
```

```python
y=dataset.iloc[:,-1]
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

X_train, X_val, y_train, y_val  = train_test_split(X_train, y_train, test_size=0.25, random_state=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knnmodel = KNeighborsClassifier(n_neighbors=3)
knnmodel.fit(X_train, y_train)
y_predict1 = knnmodel.predict(X_val)
y_predict1
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(y_val.values, y_predict1)
```

```python
acc
```

```python
y_predict2= knnmodel.predict(X_test)
```

```python
acctest = accuracy_score(y_test.values, y_predict2)
```
